```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)
```

```python
import os
path = "/content/drive/MyDrive/기계학습"
os.listdir(path)
```

```python
df = pd.read_csv(path + '/penguins_size.csv')
```

```python
numeric_columns = df.select_dtypes(include=['int64', 'float64']).columns.tolist() # 수치형 컬럼
tf_columns = df.select_dtypes(include=['object']).columns.tolist() # 범주형 컬럼

# 히스토그램을 그릴 서브플롯 생성
plt.figure(figsize=(15, 15))
for i, column in enumerate(numeric_columns, start=1):
    plt.subplot(5, 5, i)
    plt.hist(df[column], bins=20, edgecolor='black')
    plt.title(column)
    plt.tight_layout()

plt.show()
```

```python
# 히스토그램을 그릴 서브플롯 생성
plt.figure(figsize=(15, 15))
for i, column in enumerate(numeric_columns, start=1):
    plt.subplot(5, 5, i)
    plt.boxplot(df[column].dropna())
    plt.title(column)
    plt.tight_layout()

plt.show()
```

```python
# 바 플롯 그릴 서브플롯 생성
plt.figure(figsize=(15, 15))
for i, column in enumerate(tf_columns, start=1):
    plt.subplot(5, 5, i)  # Adjust the subplot grid as needed
    sns.countplot(x=column, data=df)
    plt.title(f'Countplot of {column}')
    plt.xlabel(column)
    plt.ylabel('Count')

plt.tight_layout()
plt.show()
```

```python
import seaborn as sns

# 수치형 변수 간의 산점도 그리기
sns.pairplot(df[numeric_columns])
plt.show()
```
